```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
import random 
from tqdm import tqdm
from copy import copy
```

```python
coef=[1, 2, 3]

# Тестовая квадратичная функция.
def simpleFunc(x):
    return sum((x[i]-c)**2 for i, c in enumerate(coef))

x0 = np.array([1.3, 0.7, 0.8])
#rmin = minimize(simpleFunc, x0, method='CG', options={'disp': True})
rmin = minimize(simpleFunc, x0, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```

```python
coef=[1, 2, 3]
res, points=[], []

# Тестовая функция, которая запоминает пройденные точки для отображения.
def simpleFunc2(x):
    points.append(x)
    r=sum((x[i]-c)**2 for i, c in enumerate(coef))
    res.append(r)
    return r

```

```python
# Функция отрисовки пройденного пути.
def drawPoints3D(points, res, title):
    points2, res2=np.array(points), np.array(res)
    fig = plt.figure(1, figsize=(8, 6))
    ax = Axes3D(fig, elev=-150, azim=110)
    ax.scatter(points2[:, 0], points2[:, 1], points2[:, 2], c=res2, \
               cmap=plt.cm.Set1, edgecolor='k', s=40)
    ax.set_title(title)

    plt.show()
```

```python
# Оптимизируем еще раз, отрисовываем пройденный маршрут оптимизации.
x0 = np.array([1.3, 0.7, 0.8])
res, points=[], []
#rmin = minimize(simpleFunc2, x0, method='CG')
rmin = minimize(simpleFunc2, x0, method='nelder-mead', options={'xtol': 1e-8})
print(rmin)
print("Iterantions done:", len(points))
drawPoints3D(points, res, "Nelder-Mead")
```

```python
# Функция откланяется в случайную сторону по всем координатам в заданном интервале.
# Отклонение одинаково для всех координат.
def myRandomSearch(func, x0, iterations = 100, delta = 0.1):
    res.append(func(x0))
    points.append(x0)
    x = copy(x0)
    for i in range(iterations):
        for j in range(len(x0)):
            x[j] += random.random() * 2 * delta - delta
        #print(x)
        res.append(func(x))
        points.append(copy(x))
    return res[-1], x
        
res, points=[], []
print(myRandomSearch(simpleFunc, x0, 250))
drawPoints3D(points, res, "Random Changes")
```

```python
# Функция откланяется в случайную сторону по всем координатам в заданном интервале.
# Отклонение одинаково для всех координат.
def myRandomSearch2(func, x0, iterations = 100, delta = 0.1):
    res.append(func(x0))
    points.append(x0)
    x = copy(x0)
    r_prev = 1e32
    for i in range(iterations):
        x1 = copy(x)
        for j in range(len(x0)):
            x1[j] += random.random() * 2 * delta - delta
        #print(x1)
        r = func(x1)
        #print(x1, r, r_prev)
        if r <= r_prev:
            r_prev = r
            x = x1
        res.append(r)
        points.append(copy(x))
    return r_prev, x
        
res, points=[], []
print(myRandomSearch2(simpleFunc, x0, 250))
drawPoints3D(points, res, "Random Bounded Changes")
```

```python
# Функция откланяется в случайную сторону по всем координатам в заданном интервале.
# Отклонение происходит с вероятностью 50%, отклонение одинаково по всем координатам.
def myRandomSearch3(func, x0, iterations = 100, delta = 0.1):
    res.append(func(x0))
    points.append(x0)
    x = copy(x0)
    for i in range(iterations):
        for j in range(len(x0)):
            if random.random()>0.5:
                x[j] += random.random() * 2 * delta - delta
        #print(x)
        res.append(func(x))
        points.append(copy(x))
    return res[-1], x
        
res, points=[], []
print(myRandomSearch3(simpleFunc, x0, 250))
drawPoints3D(points, res, "Stochastic Random Changes")
```

```python
# Функция откланяется в случайную сторону по всем координатам в заданном интервале.
# Отклонение происходит с вероятностью 50%, отклонение одинаково по всем координатам.
# Отклонение запоминается только при улучшении.
def myRandomSearch4(func, x0, iterations = 100, delta = 0.1, proba = 0.5):
    res.append(func(x0))
    points.append(x0)
    x = copy(x0)
    r_prev = 1e32
    for i in range(iterations):
        x1 = x
        for j in range(len(x0)):
            if random.random() > 0.5:
                x1[j] += random.random() * 2 * delta - delta
        #print(x)
        r = func(x1)
        if r <= r_prev:
            r_prev = r
            x = x1
        res.append(r)
        points.append(copy(x))
    return r_prev, x
        
res, points=[], []
print(myRandomSearch4(simpleFunc, x0, 250))
drawPoints3D(points, res, "Stochastic Bounded Random Changes")
```

Попробуем решить следующую задачу.<br>
Имеются тексты на пяти языках. Посчитаем статистику встречаемости букв на первом миллионе символов. Далее напишем функцию, которая по методу наименьших квадратов будет определять принадлежность нового текста к одному из языков.<br>

```python
from collections import Counter

""" Приводит статистику встречаемости stat к относительным значениям.
    stat - словарь, в котором ключами являются буквы, а значениями - частоты.
"""
def normalize(stat):
    s = sum(stat[c] for c in stat.keys())
    return {x : stat[x] / s for x in stat.keys()}

""" Функция определения языка. Считает сумму квадратов отклонений от заранее посчитанных
    распределений для языков. Выбирает язык, для которого сумма минимальна.
    
    text - текст, для которого определяется язык.
    stats - заранее посчитанные распределения.
    
    Возвращает номер языка и значение суммы.
"""
def detectLang(text, stats):
    maxPos = -1
    maxVal = 1e32
    stat = normalize(Counter(text))
    for i, s in enumerate(stats):
        e = sum((s.get(c, 0) - stat[c]) ** 2 for c in stat.keys())
        if e < maxVal:
            maxVal = e
            maxPos = i
    return maxPos, maxVal

""" Находит символ c в списке символов symb в интервале от s до f.
    Возвращает частоту символа или 0, если символ не найден.
"""
def findFreq(c, symb, freq, s, f):
    for i in range(s, f):
        if c == symb[i]:
            return freq[i]
    return 0
    
""" Определяет язык для текста text по частотам символов.
    text - входной текст, для которого определяется язык.
    symbs - набор символов для всех языков.
    freqs - частоты символов.
    positions - интервал, в котором находятся символы и частоты для данного языка.
"""    
def detectLang2(text, symbs, freqs, positions):
    maxPos = -1
    maxVal = 1e32
    stat = normalize(Counter(text))
    for i in range(len(positions) - 1):
        e = sum((findFreq(c, symbs, freqs, positions[i], positions[i + 1]) - stat[c]) ** 2 for c in stat.keys())
#        print(e, maxVal, positions[i], positions[i+1])
        if e < maxVal:
            maxVal = e
            maxPos = i
    return maxPos, maxVal

# Берем файлы с французскими и испанскими новостями, русской фэнтези, англлийским журналом по ИИ, немецкими новостями.
path = '/home/edward/projects/mc&s/utf8/'
filenames = ['le_parisien1_utf8.txt', 'abc-3-es_utf8.txt', 'orcs_and_dwarfs_utf8.txt', 'JMLR.txt', 'bild_utf_8.txt']
files = []
charstats = []

# Считываем первый миллион символов из каждого файла, считаем частоты символов.
for name in filenames:
    file = open(path+name)
    stat = Counter(file.read(1000000))
    stat = normalize(stat)
    files.append(file)
    charstats.append(stat)
#print(charstats[2])

# Считываем из каждого файла следующие 1000 символов, определяем для него язык.
# Просто чтобы посмотреть, что работает.
for i, f in enumerate(files):
    t = f.read(1000)
    n, v = detectLang(t, charstats)
    if n != i:
        print('Negative: ', n, v)
    else:
        print('Positive: ', n, v)

evtexts = []
chars = []
vals = []
positions = [0]

# Для каждого файла считываем 100 фрагментов по 1000 символов.
# Подготавливаем списки символов и частот, запоминаем позиции для языков.
for i, f in enumerate(files):
    evtext = []
    for j in range(100):
        evtext.append(f.read(1000))
    evtexts.append(evtext)
    for c in charstats[i].keys():
        chars.append(c)
        vals.append(charstats[i][c])
    positions.append(positions[i] + len(list(charstats[i].keys())))
    
""" Функция для оценки точности работы метода. 
"""
def evaluate(freqs):
    mistakes = 0
    ra = []
    for i, l in enumerate(evtexts):
        mistakes_cl = 0
        for t in l:
            n, v = detectLang2(t, chars, freqs, positions)
#            print(n, i, v)
            if n != i:
                mistakes += 1
                mistakes_cl += 1
        ra.append(mistakes_cl)
        
    ra.append(mistakes)
#     res_all.append(ra)
    print(mistakes)
    return mistakes
    
evaluate(vals)
```

## Функции потерь

Пусть $y_i$ - это i-е рассчитанное значение, а $\hat{y_i}$ - это i-е истинное значение. Тогда можно определить следующие функции для оценки точности результатов.

MAE (Mean Absolute Error) - средняя абсолютная ошибка $MAE=\frac{1}{N}\sum \limits_{i=1}^{N}|y_i-\hat{y_i}| $

MAPE - (Mean Absolute Percentage Error) - средняя абсолютная ошибка в процентах $MAPE=\frac{1}{N}\sum \limits_{i=1}^{N}|\frac{y_i-\hat{y_i}}{\hat{y_i}}| $

MSE (Mean Square Error) - среднеквадратичная ошибка $MSE=\frac{1}{N}\sum \limits_{i=1}^{N}(y_i-\hat{y_i})^2$
    
RMSE (Root Mean Square Error) - корень из среднекваадратичной ошибки $MSE=\sqrt{MSE}=\sqrt{\frac{1}{N}\sum \limits_{i=1}^{N}(y_i-\hat{y_i})^2}$

RMSLE (Root Mean Square Logarithmic Error) - логарифмическая среднеквадратичная ошибка $RMSLE=\sqrt{\frac{1}{N}\sum \limits_{i=1}^{N}(\log(y_i+1)-\log(\hat{y_i}+1))^2}=\sqrt{\frac{1}{N}\sum \limits_{i=1}^{N}\log^2\left(\frac{y_i+1}{\hat{y_i}+1}\right)}$

RMSPE (Root Mean Square Percentage Error) $RMSPE = \sqrt{MSPE}=\sqrt{\frac{1}{N}\sum \limits_{i=1}^{N}(\frac{y_i-\hat{y_i}}{\hat{y_i}})^2}$
